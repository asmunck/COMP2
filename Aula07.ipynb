# 1. Arquivos

## 1.1. Tipos de arquivos

Você já parou pra pensar como suas fotos e músicas ficam guardados no celular? **Cada uma fica num arquivo**. Um arquivo nada mais é do que **uma sequência de bits**. Um bit é a menor unidade de informação no computador, significa **dígito binário** (também pedacinho, em inglês) e pode ter os valores 0 ou 1. Para facilitar a parte eletrônica, os primeiros computadores agrupavam os dados em **conjuntos de 8 bits, chamados de bytes**. Cada byte pode armazenar $2^8 = 256$ valores distintos. Por convenção, **um quilobyte** (KB) possui 1024 bytes. **Um megabyte** (MB) possui 1024 KB. **Um gigabyte** (GB) possui 1024 MB. Celulares atuais possuem entre 64 e 256 GB de memória. Jogos modernos ocupam mais de 50 GB de memória. Hoje em dia, notebooks devem ter 512 GB ou 1 Terabyte (= 1024 GB) de SSD, para ter maior durabilidade.

Mas o que significa o número binário 0110.0001? Convertendo para inteiro (a calculadora do Windows faz isso), obtemos 97. E se associarmos um número a cada letra do alfabeto?

| Número | Caractere |
|:------:|:---------:|
|   97   |     a     |
|   98   |     b     |
|   99   |     c     |
|   ...  |    ...    |
|   122  |     z     |

Assim, sempre que o editor de textos ler o número 97, este será interpretado como o caractere "a". O Python tem as funções `ord` e `chr` que mostra o número associado a um caractere e vice-versa.

````python
ord('a')  # Qual é o número associado a 'a'?
````

````python
chr(97)  # Qual é o caractere associado a 97?
````

## 1.2. Escrevendo arquivos de texto

Arquivos de texto são bem simples de manipular. Você deve começar abrindo o arquivo com a função `open`. Ela retorna **um objeto que será usado para acessar o arquivo**. O método `write` escreve algo no arquivo. Quando terminar de usar o arquivo, **nunca esqueça de fechá-lo** usando o método `close`. Veja o exemplo abaixo:

````python
a = open("teste.txt", 'w')

a.write("Isto é apenas um teste.\n")
a.write("Seria isso apenas um teste?")

a.close() # IMPORTANTE!
````

Depois de rodar a célula acima, clique no ícone de arquivo &#128448; na barra de atalhos à esquerda. **Veja que o arquivo teste.txt está lá!** Clique duas vezes nele para exibir o conteúdo. Está certinho né? O arquivo não existia **mas foi criado pela função** `open` na memória disponibilizada pelo Colab **para esta sessão**. Ou seja, quando você fechar o notebook, o arquivo sumirá 😱!

**O arquivo foi criado para escrita**, por isso a opção `'w'` foi usada no segundo argumento de `open`. O primeiro argumento é o nome do arquivo e usamos uma variável `a` (de arquivo) para guardar o objeto. Veja outras opções de abertura:

| Opção | Modo                                                                                    |
|:-----:|-----------------------------------------------------------------------------------------|
|  'r'  | Abre um arquivo para leitura (read). É o valor padrão para o modo.                      |
|  'w'  | Abre para escrita (write). Se já existe, apaga o conteúdo anterior.                     |
|  'a'  | Abre um arquivo para anexar. Começa a escrever no fim dele.                             |
|  'x'  | Cria um arquivo novo para escrever. Gera erro se o arquivo já existe. |
|  '+'  | Usado com 'r' ou 'w' para ler e escrever no mesmo arquivo. |
|  'b'  | Usado com as opções acima para manipular arquivos binários. |

Note o `\n` ao final da primeira string. Ele foi adicionado para indicar que se deve ir para a próxima linha do texto (n é de *new line*). O método `write` escreve exatamente o que lhe foi passado. O método `close` é muito importante para que todos os dados sejam gravados no arquivo e a conexão com ele seja encerrada.

**Como manter o arquivo na memória?**

* O arquivo pode ser baixado no seu computador.
* É possível conectar o Colab ao seu Google Drive. Dessa forma os arquivos modificados não serão apagados ao final da sessão.
* Usando o Python no seu computador (com o IDLE) ou celular, os arquivos serão criados no seu próprio HD ou SSD.

## 1.3. Lendo arquivos de texto

Ler arquivos de texto também é muito simples. O método `read` lê todo o conteúdo do arquivo de uma vez só. Se o conteúdo for muito grande, é possível ler um linha de cada vez usando `readline`. O exemplo a seguir lê o conteúdo do arquivo "README.md" que está no diretório "sample_data". Este diretório e seus arquivos são criados automaticamente pelo Colab.

````python
# read lê todo o conteúdo de uma vez só
f = open("sample_data/README.md")
todo_texto = f.read()
print(todo_texto)
f.close()
````

````python
# Enquanto readline lê apenas uma linha por vez
f = open("teste.txt")
print(f.readline())
print(f.readline())
f.close()
````

Porquê tem uma linha em branco no meio? Porque `readline` retorna uma linha do arquivo incluindo o último caractere `\n`, de nova linha, mas o `print` sempre adiciona um `\n` quando imprime. Este segundo `\n` não imprime nada, só deixa uma linha em branco. Uma forma evitar isso é usar o método `str.strip` para remover caracteres `\n`.

````python
f = open("teste.txt")
print(f.readline().strip('\n'))
print(f.readline().strip('\n'))
f.close()
````

Uma forma interessante de olhar linha por linha é **usar o objeto de arquivo como iterável** no laço for. Neste exemplo, usamos outra estratégia para não deixar linhas em branco. Você sabe dizer qual?

````python
arquivo = open("teste.txt")

for linha in arquivo:
  # Cada iteração fornece uma linha do arquivo
  # (como se fosse: linha = arquivo.readline())
  print(linha, end ='')

f.close()
````

## 1.4. Passei a noite procurando tu...

Quando `readline` lê uma linha do arquivo, a posição atual é guardada. Caso contrário, não teria como saber onde continuar a ler certo? Essa posição é chamada de **ponteiro** e conta caracteres começando em zero. Podemos modificar o ponteiro usando o método `seek`. Vamos praticar fazendo alguns "pulos" no arquivo:

````python
f = open("teste.txt")
print(f.readline().strip('\n'))
f.seek(0)
print(f.readline().strip('\n'))
f.seek(0)
print(f.readline().strip('\n'))
f.close()
````

## 1.5. Nunca esquecer de fechar!

A linguagem Python é cheia de surpresas! Tem uma forma de tornar a operação com arquivos mais segura, coloca-se todas as operações de arquivo dentro de um bloco começando com `with open(<nome>, <modo>) as <variável>:` onde `variável` é o nome da variável que guarda o objeto arquivo. Veja o exemplo.

````python
with open("teste2.txt") as tarefas:
  print(tarefas.readline())
  print(tarefas.readline())
  print(tarefas.readline())



# Sentiu falta de tarefas.close()?
````

Quando o bloco `with` termina, **o arquivo é automaticamente fechado**. Assim é fácil né?

# 2. Banco de dados

Hoje em dia se fala muito em **ciência de dados**. A ciência de dados usa métodos estatísticos para analisar uma grande quantidade de dados e fazer previsões. Às vezes, estes dados aparecem como arquivos **CSV** (*Comma Separated Values*), que é um arquivo de texto com uma tabela de valores separados por vírgula. Outras vezes aparecem na forma de **banco de dados** que são armazenados em servidores próprios, locais ou na nuvem, e dependem de algum módulo para interagir com os dados.

## 2.1. Banco de dados SQL

**SQL** significa *Structured Query Language*. É **uma linguagem padronizada para acessar e manipular bancos de dados**. O banco de dados é dito **estruturado** porque está organizado em tabelas. Grandes empresas como a Google, IBM e Amazon oferecem serviços pagos para armazenar banco de dados.

O Python tem um módulo chamado `sqlite3` para criar e manipular bancos locais (salvos no próprio computador). Infelizmente, não sei se teremos tempo para olhar esse módulo.

## 2.2 Banco de dados NoSQL

**NoSQL** são bancos de dados não estruturados que **organizam seus dados na forma de entradas**. Em geral, são dados que não precisam ser colocados em tabelas pois não são padronizados. **Cada entrada é um dicionário** com quantas chaves forem necessárias. Há serviços populares de bancos NoSQL na nuvem, como o MongoDB.

Quando for pequeno, podemos gravar um banco NoSQL no **formato JSON**, que é um arquivo de texto comumente usado para transferência de dados em servidores web. Para saber como ele se parece, abra o arquivo "sample_data/anscombe.json". **É uma grande lista de dicionários certo?**

Estude o exemplo a seguir, para saber como usar módulo `json` para abrir um arquivo JSON.

````python
import json
with open("sample_data/anscombe.json") as a:
  dados = json.load(a)

# Agora dados é uma lista com todos os dicionários
# Ver o primeiro da lista
dados[0]
````

````python
# Cada conteúdo do dicionário pode ser acessado pela chave
# Por exemplo, acessar o X da entrada 10
x = dados[10]['X']
print(x)
````

Repare que o próprio objeto de arquivo foi passado como argumento da função `json.load`. Faremos a mesma coisa, para gravar a lista com `json.dump`. Isso depois de adicionar a novíssima "série V" com coordenadas 3D.

````python
# Adiciona a nova série de dados
dados.append( dict(Series='V', X=1.0, Y=2.0, Z=3.0) )
dados.append( dict(Series='V', X=3.0, Y=1.0, Z=2.0) )
dados.append( dict(Series='V', X=2.0, Y=3.0, Z=1.0) )

# Agora vamos salvar no arquivo dados.json (não esquecer do modo 'w')
with open("novo.json", 'w') as a:
  json.dump(dados, a, indent=2)
````

Agora abra o arquivo "novo.json". Repare que as novas entradas estão lá no final!

```
  "Frohike": {

  },
```
Não tem problema esse dicionário não ter a chave ramal pois, como vimos, a entradas não são padronizadas. Em seguida, escreva um código para recuperar o dicionário do arquivo e imprimir o conteúdo da chave "Frohike".



````python
# Seu código aqui
import json

funcionario = {
    "Frohike": {
        "cargo": "programador",
        "idade": 66,
        "salario": 21000
    }
}

with open('funcionarios.json', 'r') as file:
    conteudo = json.load(file)

conteudo.update(funcionario)

with open('funcionarios.json', 'w') as file:
    json.dump(conteudo, file)

with open('funcionarios.json', 'r') as file:
    conteudo_atualizado = json.load(file)

print(json.dumps(conteudo_atualizado, indent=2))
````
