b) **[Criação de Matriz Especial]**: Escreva uma função chamada cria_matriz_triangular_superior que recebe um inteiro *n* como parâmetro e retorna uma matriz triangular superior de ordem $n$ usando NumPy.

```python
import numpy as np

def cria_matriz_triangular_superior(n):
  matriz = np.zeros((n, n))

  for s in range(n):
    for j in range(s, n):
      matriz[s][j] = 1

  return matriz

print(cria_matriz_triangular_superior(3))
print()
print(cria_matriz_triangular_superior(5))
```

c) **[Criação de Matriz Especial]**: Escreva uma função chamada cria_matriz_triangular_inferior que recebe um inteiro *n* como parâmetro e retorna uma matriz triangular inferior de ordem $n$ usando NumPy.

```python
import numpy as np

def cria_matriz_triangular_inferior(n):
    matriz = np.zeros((n, n))
    for j in range(n):
        for i in range(j, n):
            matriz[i][j] = 1
    return matriz

cria_matriz_triangular_inferior(5)
```

A função de transferência de um sistema de segunda ordem é dada pela seguinte equação:
$$
H = 1 / (1 + 2 \zeta \omega_n  1j - \omega_n^2 t),
$$
onde $\omega_n$ é a frequência natural não amortecida, $\zeta$ é a razão de amortecimento e $t$ é o tempo.

Dados, $\omega_n = 2.0$ e $\zeta = 0.5$:

a) Crie um vetor de tempo t (*array NumPy*), com 1000 pontos igualmente espaçados no intervalo $[0,10]$.

b) Calcule a função de transferência H de um sistema de segunda ordem como uma função do tempo e armazene o resultado em um array NumPy.

### Exercício 3
Faça uma função que recebe uma lista de valores e o número de linhas n e verifica se essa lista forma
uma matriz quadrada n x n.

Caso forme,

  a) converta a lista para um numpy array,

  b) altere o seu formato para ser igual a n x n e
  
  c) **retorne** a matriz e o seu determinante.
  
**Dica**: Utilize a biblioteca **NumPy** e o módulo **linalg** da biblioteca **Scipy**.

Caso não forme, **retorne** uma mensagem informando ao usuário.


```python
import numpy as np
from scipy.linalg import det

def matriz_quadrada(valores, n):
    if len(valores) != n*n:
        return "Valores inválidos para preencher uma matriz quadrada."
    matriz = np.array(valores).reshape(n, n)
    determinante = det(matriz)
    return matriz, determinante    

matriz_quadrada([1, 2, 3, 4], 2)

```

### Exercício 5:
O código abaixo representa uma interface gráfica utilizando os módulos Tkinter e Numpy. Explique, brevemente, como será a interface e quais componentes estão presentes nela. Ao final, informe qual será o resultado quando o usuário clicar no único botão da interface.

```python
import numpy as np
from tkinter import *

class Janela(object):

    def __init__(self, toplevel):
        self.resultado = 0
        self.toplevel = toplevel

        self.text = StringVar()
        self.text.set("Resultado: " + str(self.resultado))
        self.label = Label(self.toplevel, textvariable = self.text)
        self.label.pack(side = TOP)

        self.botao = Button(self.toplevel, text="Calcular")
        self.botao.pack(side = TOP)
        self.botao.bind("<Button-1>", self.calcula)

    def calcula(self, event):
        A = np.arange(9).reshape(3, 3)
        B = np.array([[5, 10], [20, 40], [2, 4]])
        self.resultado = np.dot(A,B)
        self.text.set("Resultado: " + str(self.resultado))

root = Tk()
Janela(root)
root.mainloop()
```

Resposta: 
O código define uma classe Janela que é herdeira de object, com o método Tkinter e Numpy, a janela construida armazenou os valores do produto de duas matrizes pré definidas na função calcula. No topo é definido a StringVar, que armazena a variável self.text e recebe como resultado a variável realizada na função calcula, iniciando com 0 por padrão. O botão recebe o "<'Button-1'> como comando para a ativação da função, que é o botão esquerdo do mouse, retornando o produto das matrizes.

### Exercício 6:
Implemente um jogo de forca simplificado seguindo os passos a seguir:

a) Crie uma classe chamada `Forca`, filha da classe `Frame`

b) A palavra a ser advinhada deve ser armazenada em um arquivo chamado ”palavra.txt” que ficará na mesma pasta do código. Ao iniciar a aplicação, a palavra deverá ser carregada do arquivo para
um atributo.

c) A janela deve possuir 3 `widgets`: 2 do tipo `Label` e 1 do tipo `Entry` conforme mostra abaixo:
exercicio6-P2.png

<!-- <center>
    <div>
        <img src= "Figs/exercicio6-P2.png" width="300"/>
    </div>

```python
import tkinter as tk
from tkinter import Frame, Label, Entry
import random

class Forca(Frame):
    def __init__(self, mestre=None):
        super().__init__(mestre)
        self.mestre = mestre
        self.pack()
        self.criar_widgets()

    def criar_widgets(self):
        self.rotulo_palavra = Label(self)
        self.rotulo_palavra["text"] = "Adivinhe a palavra"
        self.rotulo_palavra.pack(side="top")
        self.rotulo_palavra_adivinhada = Label(self)
        self.rotulo_palavra_adivinhada.pack(side="top")
        self.entrada = Entry(self)
        self.entrada.bind('<Return>', self.verificar_palpite)
        self.entrada.pack(side="bottom")

    def carregar_palavra(self):
        with open("palavra.txt", "r") as arquivo:
            self.palavras = [palavra.strip().lower() for palavra in arquivo.readlines()]
            self.palavra = random.choice(self.palavras)
            self.palavra_adivinhada = ["-"] * len(self.palavra)
            self.atualizar_rotulo_palavra_adivinhada()

    def atualizar_rotulo_palavra_adivinhada(self):
        self.rotulo_palavra_adivinhada["text"] = "".join(self.palavra_adivinhada)

    def verificar_palpite(self, evento):
        palpite = self.entrada.get().lower()
        self.entrada.delete(0, 'end')
        for i, letra in enumerate(self.palavra):
            if letra == palpite:
                self.palavra_adivinhada[i] = palpite
        self.atualizar_rotulo_palavra_adivinhada()

raiz = tk.Tk()
app = Forca(mestre=raiz)
app.carregar_palavra()
app.mainloop()
```

```python
import pandas as pd
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Criar um DataFrame hipotético com dados de produção
data = {'Tempo': np.array([0, 1, 2, 3, 4, 5]),  # Meses
        'Producao': np.array([100, 120, 90, 80, 70, 60])}  # Barris por mês

df = pd.DataFrame(data)

# Função que descreve a produção em relação ao tempo
def production_function(t):
    # Usando uma função fictícia para ilustração
    return 100 - 10 * t

# Função para calcular a integral da produção em relação ao tempo
def integrate_production(t):
    integral_result, _ = quad(production_function, 0, t)
    return integral_result

# Aplicar a função de integração ao DataFrame
df['Volume_Total'] = df['Tempo'].apply(integrate_production)

# Exibir o DataFrame resultante
print(df)

# Plotar gráfico de produção ao longo do tempo
plt.plot(df['Tempo'], df['Producao'], label='Produção Mensal')
plt.xlabel('Tempo (meses)')
plt.ylabel('Produção (barris por mês)')
plt.title('Produção Mensal')
plt.legend()
plt.show()

# Plotar gráfico do volume total acumulado
plt.plot(df['Tempo'], df['Volume_Total'], color = 'orange', label='Volume Total Acumulado')
plt.xlabel('Tempo (meses)')
plt.ylabel('Volume Total Acumulado')
plt.title('Volume Total Acumulado de Petróleo ao Longo do Tempo')
plt.legend()
plt.show()
```

### Exercício 8:
**[Persistência de dados]**: Desenvolva um programa para gerenciar uma lista de compras. Cada item da lista deve ter um nome, quantidade e status (comprado ou não). Utilize um arquivo para persistir esses dados. O programa deve oferecer as seguintes funcionalidades:

a) Adição de novos itens à lista.

b) Listagem de todos os itens da lista.

c) Remoção de itens da lista.

```python
import json

class ListadeCompras:
    def __init__(self, arquivo): 
        self.arquivo = arquivo
        try:
            with open(self.arquivo, 'r') as file:
                self.lista = json.load(file)
        except FileNotFoundError:
            self.lista = []
        except json.JSONDecodeError:
            self.lista = []
            
    def adicionar_item(self, nome, quantidade, status):
        item = {'nome': nome, 'quantidade': quantidade, 'status': status}
        self.lista.append(item)
        self.salvar()
    
    def listar_itens(self):
        for item in self.lista:
            print(item['nome'], item['quantidade'], item['status'])

    def remover_item(self, nome):
        self.lista = [item for item in self.lista if item['nome'] != nome]
        self.salvar()

    def salvar(self):
        with open(self.arquivo, 'w') as file:
            json.dump(self.lista, file)
        
        
```

```python
lista = ListadeCompras('lista_compras.json')
lista.adicionar_item('Papel Toalha | ', 2, ' | Não Comprado')
lista.adicionar_item('Sabonete | ', 3, ' | Comprado')
lista.adicionar_item('Pão | ', 5, ' |Não Comprado')
lista.listar_itens()
lista.remover_item('Papel Toalha')

```
