```python
def ler_artilheiros():
    """essa funcao tem como objetivo receber através do input os dados dos jogadores, como nome, rodada e gols. 
    caso seja digitado fim no input o while é quebrado e então os próximos input são chamados, dessa forma, 
    os dados são adicionados a uma matriz definida inicialmente como artilheiros e valor zero."""
    artilheiros = []
    while True:
        nome = input("Qual é o nome do jogador? Alternativamente, digite 'fim' para parar a contagem.: ")
        if nome.lower() == 'fim':
            break
        rodada = input("Digite a rodada: ")
        gols = int(input("Digite o número de gols: "))
        artilheiros.append([nome, rodada, gols])
    return artilheiros

def media_gols(artilheiros):
    """nessa funcao criamos uma nova lista a qual servira para fazer o calculo da media de gols de cada jogador,
    sendo necessario converter a string para valores inteiros para a operação matematica"""
    media_gols = []
    for jogador in artilheiros:
        nome, rodada, gols = jogador
        if rodada == 'classificatória':
            jogos = 3
        elif rodada == 'oitavas':
            jogos = 4
        elif rodada == 'quartas':
            jogos = 5
        elif rodada == 'semi':
            jogos = 6
        elif rodada == 'final':
            jogos = 7
        media = gols / jogos
        media_gols.append([nome, media])
    return media_gols

def main():
    """nessa funcao main, chamamos as outras duas definidas anteriormente, na qual a primeira ativa a funcao que pede os jogadores, as rodadas e os seus gols, e em seguida,
    ativa a funcao media para calcular a media de gols, por último, definimos o vencedor como aquele com a maior média na lista, o argumento 0 da lista e o nome do jogador,
    e o 1 e a media de gols."""
    artilheiros = ler_artilheiros()
    media = media_gols(artilheiros)
    vencedor = max(media, key=lambda x: x[1])
    print(f"O vencedor da Chuteira de Ouro é {vencedor[0]} com uma média de {vencedor[1]} gols por jogo.")

ler_artilheiros()
```

```python
import random

# Gerando a matriz de números
modelo = [[0] * 4 for _ in range(4)]
jogo = [[0] * 4 for _ in range(4)]

numero = 1
while numero < 9:
    i1, j1 = random.randint(0, 3), random.randint(0, 3)
    i2, j2 = random.randint(0, 3), random.randint(0, 3)
    
    if modelo[i1][j1] == 0 and modelo[i2][j2] == 0 and (i1 != i2 or j1 != j2):
        modelo[i1][j1] = modelo[i2][j2] = numero
        numero += 1

# Mostrando a matriz escondida
def mostrar_jogo():
    for linha in jogo:
        for elemento in linha:
            if elemento == 0:
                print('*', end=' ')
            else:
                print(elemento, end=' ')
        print()

# primeiro turno
def posicao_valida(x, y):
    return 0 <= x < 4 and 0 <= y < 4 and jogo[x][y] == 0

def joga_turno():
    mostrar_jogo()
    
    x1, y1 = map(int, input("Digite a primeira posição (linha coluna): ").split())
    while not posicao_valida(x1, y1):
        print("Posição inválida!")
        x1, y1 = map(int, input("Digite a primeira posição (linha coluna): ").split())
    
    x2, y2 = map(int, input("Digite a segunda posição (linha coluna): ").split())
    while not posicao_valida(x2, y2):
        print("Posição inválida!")
        x2, y2 = map(int, input("Digite a segunda posição (linha coluna): ").split())
    
    if modelo[x1][y1] == modelo[x2][y2]:
        print("Parabéns, você acertou!")
    else:
        print("Você errou... tente novamente.")
        jogo[x1][y1] = jogo[x2][y2] = 0
    mostrar_jogo()

def jogo_memoria():
    while any(0 in linha for linha in jogo):
        joga_turno()

    print("Parabéns, você venceu!")

jogo_memoria()
```
