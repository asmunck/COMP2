# 1. O que é o NumPy?

O **NumPy** (Numerical Python) é um pacote fundamental para computação científica e numérica. Ele fornece suporte para *arrays* *n*-dimensional (ou seja, matrizes e tensores) e funções matemáticas de alto desempenho para operações em massa em dados numéricos. Os *arrays* devem conter todos os elementos do mesmo tipo, ou seja, devem ser homogêneos.

**NumPy** é uma das bibliotecas mais importantes para cientistas de dados, engenheiros e pesquisadores que trabalham com computação numérica e análise de dados.

Faça uma função *gera_matriz* que recebe como parâmetros de entrada *(a,b,n,m)* e que tem como retorno um array NumPy de formato $m \times m$, onde os elementos são os números igualmente espaçados obtidos ao se dividir o intervalo [a,b]. Caso não seja possível construir tal array, a função deve imprimir a mensagem "Nao e possivel construir".

```
# Exemplo de uso:
a = 0
b = 10
n = 16
m = 4

resultado_1 = gera_matriz(a, b, n, m)
# Saída
[[ 0.          0.66666667  1.33333333  2.        ]
 [ 2.66666667  3.33333333  4.          4.66666667]
 [ 5.33333333  6.          6.66666667  7.33333333]
 [ 8.          8.66666667  9.33333333 10.        ]]

resultado_2 = gera_matriz(0, 10, 10, 4)
# Saída
Nao e possivel construir
```




````python
# Escreva sua resposta aqui!
import numpy as np

a = 1
b = 16
n = 4
m = 4

def gera_matriz(a, b, n, m):
    if n * m != (b - a + 1):
        print("Não é possível construir a matriz.")
    numeros = np.arange(a, b + 1)

    try:
        matriz = numeros.reshape(m, m)
        return matriz
    except ValueError:
        print("Não é possível construir.")

resultado = gera_matriz(a, b, n, m)
print("Matriz resultante:")
print(resultado)

````

### Exercício 7 [1pt]

A maioria dos **sistemas lineares** podem ser resolvidos na forma de equações matriciais. Por exemplo, o sistema

$$
  \begin{cases}
    a_{11} x + b_{12}\, y = c_1 \\
    a_{21} x + b_{22}\, y = c_2
  \end{cases}
$$

pode ser escrito na forma
$$
\begin{align}
  \begin{bmatrix}
    a_{11} & a_{12} \\
    a_{21} & a_{22}
  \end{bmatrix}
  \begin{bmatrix}
    x \\
    y
  \end{bmatrix}
  &=
  \begin{bmatrix}
    c_1 \\
    c_2
  \end{bmatrix}
  \\
  A\, \vec{v} &= \vec{b}
\end{align}
$$
Note que as **matrizes coluna são entendidas como vetores**.

Uma forma de encontrar os valores de $x$ e $y$ que satisfazem as duas equações do sistema é **triangularizar** a matriz $A$. Outra forma é procurar a **matriz inversa** $A^{-1}$, que tem a propriedade $A^{-1}A = I$. A matriz identidade $I$, como você deve saber, não altera a matriz que está multiplicando. Depois de encontrar a inversa, **multiplicamos os dois lados da equação** por ela, para obter
$$
  \begin{align}
    A^{-1}A\,\vec{v} &= A^{-1}\vec{b} \\
    I\,\vec{v} &= A^{-1}\vec{b} \\
    \vec{v} &= A^{-1}\vec{b}
  \end{align}
$$

Para inverter uma matriz, o NumPy tem a função `numpy.linalg.inv` ([link](https://numpy.org/doc/stable/reference/generated/numpy.linalg.inv.html) para a documentação).

**Veja um exemplo:** <br>
A solução para o sistema
$$
  \begin{cases}
    x - y = 0 \\
    x + y = 2
  \end{cases}
$$
é $x = y = 1$.

````python
from numpy.linalg import inv
A = [[1, -1], [1, 1]]
Ainv = inv(A)
c = np.array([[0], [2]])
# Solução do sistema
Ainv @ c
````

Agora resolva o seguinte problema com NumPy. Numa certa lanchonete, a tabela mostra quantos salgados e sucos foram pedidos nas mesas 1 e 2, assim como as contas. Calcule a conta da mesa 3.

|          | Mesa 1 | Mesa 2 | Mesa 3 |
|----------|:------:|:------:|:------:|
| Salgados |    3   |    5   |    1   |
| Sucos    |    2   |    3   |    2   |

# Avalie

````python
import numpy as np

salgados = np.array([3, 5, 1])
sucos = np.array([2, 3, 2])
contas = np.array([34, 54, 0])

valor_salgado = contas[0] / np.sum(salgados)
valor_suco = contas[1] / np.sum(sucos)

quantidade_salgados_mesa3 = 1
quantidade_sucos_mesa3 = 2
contas[2] = (quantidade_salgados_mesa3 * valor_salgado) + (quantidade_sucos_mesa3 * valor_suco)

print(f"A conta da Mesa 3 é R$ {contas[2]:.2f}")
````
