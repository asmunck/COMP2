## 1.3 Mudando a perspectiva

Veja três formas de somar inteiros:

````python
print( int.__add__(3, 4) )
print( (3).__add__(4) )
print( 3 + 4 )
````

1. `int.__add__(3, 4)` &dash;
Na primeira forma, temos **uma função atuando sobre argumentos passivos**.
2. `(3).__add__(4)` &dash;
A segunda forma é a perspectiva da POO, **o objeto** (número 3) **gera uma ação de soma** com outro objeto (número 4). O prefixo agora é o primeiro argumento (o `self`).
3. `3 + 4` &dash; A terceira é igual às duas anteriores, **substituindo o método por seu símbolo correspondente** (o +).

Estas formas são equivalentes, mas a gente quer facilidade certo? 😎 Então, não esqueça:
> Em POO, **uma função da classe é chamada com**<br> `<objeto>.<método>(<argumentos depois de self>)`

Veja alguns exemplos:
* `list.append(media_gols, 5)` &rarr; `media_gols.append(5)`
* `str.split('O rato roeu', 'r')` &rarr; `'O rato roeu'.split('r')`
* `str.replace(texto, '.', ';')` &rarr; `texto.replace('.', ';')`

# 2. Conjuntos

## 2.1. Criando conjuntos

Conjuntos são **uma coleção de itens sem repetição e sem ordenamento**. A razão de termos relembrado dicionários antes, é que **a declaração de conjuntos é parecida com a de dicionários**. Tome cuidado para não confundir!

> Um conjunto é definido com itens separados por vírgula, dentro de um par de chaves `{}`.

Por exemplo:

````python
# Definindo um conjunto s1
conjunto1 = {'a', 'b'}
type(conjunto1)
````

````python
# Verificando que itens não são repetidos
conjunto2 = {'a', 'b', 'a'}
print(conjunto2)
````

````python
# Conjuntos podem conter qualquer tipo imutável
conjunto3 = {'a', 'b', 1, (1.5, True)}
print(conjunto3)
````

````python
# Listas são mutáveis, por isso não entram!
conjunto4 = {'a', 'b', 1, (1.5, True), [3, 5]}
print(conjunto4)
````

## 2.2. Operações entre conjuntos

Os conjuntos possuem algumas operações matemáticas:
* Um conjunto é subconjunto do outro: $A \subset B$
* Dois conjuntos são disjuntos: $A \cap B = \emptyset$
* Interseção $A \cap B$
* União $A \cup B$
* Diferença $A - B$

Para ilustrar, vamos criar alguns conjuntos primeiro:

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
````

*As subseções de i a v têm questões valendo 0,1 pt cada.*

### i. `issubset`

Se **todos os elementos** de um conjunto $A$ estão presentes no conjunto $B$, então $A$ **está contido em** $B$. Podemos escrever $A \subset B$. Para verificar se isto é verdade, podemos usar o método `issubset`.

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
b.issubset(a)
````

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
a.issubset(b)
````

### ii. `isdisjoint`

Por outro lado, se $A$ **não tem elementos em comum com** $B$, então eles **são disjuntos**. Matematicamente, podemos escrever que sua interseção resulta num conjunto vazio, $A \cap B = \emptyset$.

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
a.isdisjoint(b)
````

````python
b.isdisjoint(c)
````

**Responda:** `c` é disjunto de `a`?

### iii. `intersection` ou `&`

A interseção entre os conjuntos $A$ e $B$ é o conjunto com os elementos que **estão em $A$ E $B$** (se estiver somente em $A$ ou $B$ não é elemento da interseção). Escreve-se $A \cap B$.

````python
print(a.intersection(b))
print(a & b)
````

**Responda:** Qual é a interseção de `c` com `a`?

````python
print(c.intersection(a))
print(a & c)
````

### iv. `union` ou `|`

A união dos conjuntos $A$ e $B$ é o conjunto que contém os elementos que **estão em $A$ OU $B$**. Escreve-se $A \cup B$.

````python
print(a.union(b))
print(a | b)
````

**Responda:** Qual é a união de `a` com `b` e `c`?

````python
# Seu código aqui
print(a.union(b, c))
print(a | b | c)
````

### v. `difference` ou `-`

A diferença entre os conjuntos $A$ e $B$ é o conjunto de elementos que **estão em $A$ mas não em $B$**.

````python
print(a.difference(b))
print(a - b)
````

**Responda:** Qual é a diferença entre `b` e `a`?

````python
# Seu código aqui
print(b.difference(a))
print(b - a)
````

## 2.4. Criando conjuntos com `set`

**set** é o nome da classe, mas também pode ser usada como **função para criar conjuntos** a partir de tuplas, listas, etc. Veja os exemplos abaixo.

````python
# Cria a partir de uma tupla
a = set((1, 2, 3))
print(a)

# Cria a partir de uma lista
b = set(['a', 'b', 'c', 'a'])
print(b)

# Cria a partir de uma string
c = set('abca')
print(c)

# Cria um conjunto vazio
d = set()
print(d)
````
