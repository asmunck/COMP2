## 1.3 Mudando a perspectiva

Veja tr√™s formas de somar inteiros:

````python
print( int.__add__(3, 4) )
print( (3).__add__(4) )
print( 3 + 4 )
````

1. `int.__add__(3, 4)` &dash;
Na primeira forma, temos **uma fun√ß√£o atuando sobre argumentos passivos**.
2. `(3).__add__(4)` &dash;
A segunda forma √© a perspectiva da POO, **o objeto** (n√∫mero 3) **gera uma a√ß√£o de soma** com outro objeto (n√∫mero 4). O prefixo agora √© o primeiro argumento (o `self`).
3. `3 + 4` &dash; A terceira √© igual √†s duas anteriores, **substituindo o m√©todo por seu s√≠mbolo correspondente** (o +).

Estas formas s√£o equivalentes, mas a gente quer facilidade certo? üòé Ent√£o, n√£o esque√ßa:
> Em POO, **uma fun√ß√£o da classe √© chamada com**<br> `<objeto>.<m√©todo>(<argumentos depois de self>)`

Veja alguns exemplos:
* `list.append(media_gols, 5)` &rarr; `media_gols.append(5)`
* `str.split('O rato roeu', 'r')` &rarr; `'O rato roeu'.split('r')`
* `str.replace(texto, '.', ';')` &rarr; `texto.replace('.', ';')`

## 2.2. Opera√ß√µes entre conjuntos

Os conjuntos possuem algumas opera√ß√µes matem√°ticas:
* Um conjunto √© subconjunto do outro: $A \subset B$
* Dois conjuntos s√£o disjuntos: $A \cap B = \emptyset$
* Interse√ß√£o $A \cap B$
* Uni√£o $A \cup B$
* Diferen√ßa $A - B$

Para ilustrar, vamos criar alguns conjuntos primeiro:

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
````

*As subse√ß√µes de i a v t√™m quest√µes valendo 0,1 pt cada.*

### i. `issubset`

Se **todos os elementos** de um conjunto $A$ est√£o presentes no conjunto $B$, ent√£o $A$ **est√° contido em** $B$. Podemos escrever $A \subset B$. Para verificar se isto √© verdade, podemos usar o m√©todo `issubset`.

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
b.issubset(a)
````

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
a.issubset(b)
````

### ii. `isdisjoint`

Por outro lado, se $A$ **n√£o tem elementos em comum com** $B$, ent√£o eles **s√£o disjuntos**. Matematicamente, podemos escrever que sua interse√ß√£o resulta num conjunto vazio, $A \cap B = \emptyset$.

````python
a = {1, 2, 3, 4, 5}
b = {1, 2, 3}
c = {10}
a.isdisjoint(b)
````

### iii. `intersection` ou `&`

A interse√ß√£o entre os conjuntos $A$ e $B$ √© o conjunto com os elementos que **est√£o em $A$ E $B$** (se estiver somente em $A$ ou $B$ n√£o √© elemento da interse√ß√£o). Escreve-se $A \cap B$.

````python
print(a.intersection(b))
print(a & b)
````

**Responda:** Qual √© a interse√ß√£o de `c` com `a`?

### iv. `union` ou `|`

A uni√£o dos conjuntos $A$ e $B$ √© o conjunto que cont√©m os elementos que **est√£o em $A$ OU $B$**. Escreve-se $A \cup B$.

````python
print(a.union(b))
print(a | b)
````

**Responda:** Qual √© a uni√£o de `a` com `b` e `c`?

### v. `difference` ou `-`

A diferen√ßa entre os conjuntos $A$ e $B$ √© o conjunto de elementos que **est√£o em $A$ mas n√£o em $B$**.

````python
print(a.difference(b))
print(a - b)
````

**Responda:** Qual √© a diferen√ßa entre `b` e `a`?

## 2.4. Criando conjuntos com `set`

**set** √© o nome da classe, mas tamb√©m pode ser usada como **fun√ß√£o para criar conjuntos** a partir de tuplas, listas, etc. Veja os exemplos abaixo.

````python
# Cria a partir de uma tupla
a = set((1, 2, 3))
print(a)

# Cria a partir de uma lista
b = set(['a', 'b', 'c', 'a'])
print(b)

# Cria a partir de uma string
c = set('abca')
print(c)

# Cria um conjunto vazio
d = set()
print(d)
````
